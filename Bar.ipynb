```python
x = 2
while x >= 2:
  print('========= Um Bar que fala sobre você =====================')
  print('1 - Beber uma garafa para encontrar')
  print('2 - Procura-lá em outra garafa')
  print('3 - desistir de procurar nessa garafa')
  print('4 - Sair do bar')
  print('========= Um Bar que fala sobre você =====================\n')
  y = int(input())
  if y == 1:
    print('\nVocê bebeu uma garafa, mas achou você mesmo\n')
  elif y == 2:
    print('\nVocê achou outra garafa, bebeu, mas não achou\n')
  elif y == 3:
    print('\nA garafa não saiu da sua mão, mas você saiu de dentro da garafa\n')
  elif y == 4:
    print('\nVocê não pode sair sem terminar de beber e encontra-lá\n')
  else:
    print('\nVocê ainda está na garafa, do mesmo bar...\n')
```
